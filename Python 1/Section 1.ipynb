# Data Types
Python has several built-in data types that are commonly used in programming. their are 5 main data types in python:
1. **String**: "____" or '____' anything inside quotes is a string.
```python
print("Hello, World!")
print(" ") 
print("1234")
```
2. **Integer**: Positve or negative whole numbers without a decimal point. e.g., 1, -5, 42



```python
print(42)
```

3. **Float**: Numbers with decimal points. e.g., 3.14, -0.001, 2.0


```python
print(42.22)
```

4. **Boolean**: Represents one of two values: True or False.
```python
print(10 > 5) # True
print(10 < 5) # False
```
5. **Complex number**: Numbers with a real and imaginary part. e.g., 3 + 4j, -1 + 0j
```python
print(3 + 5j)
```

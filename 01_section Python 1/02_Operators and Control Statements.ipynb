# Fizz - Buzz

Given an Integer as input
1. If the number is divisible by 3, print Fizz
2. If the number is divisible by 5, print Buzz
3. If the number is divisible by both 3 and 5, print FizzBuzz
4. If the number is not divisible by either 3 or 5, print the number


```python
number = int(input())
if number % 3 ==0:
    print("Fizz")
elif number % 3 ==0:
    print("Buzz")
elif (number %3 == 0) and  (number %5 ==0):
    print("Fizz-Buzz")
```
